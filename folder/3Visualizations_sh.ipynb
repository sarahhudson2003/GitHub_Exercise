```python
titanic_data = sns.load_dataset('titanic')
print(titanic_data.head())
```

```python
# use survived as target variable
proportion_table = pd.crosstab(titanic_data['alone'], titanic_data['survived'], normalize = 'index')

proportion_table.plot(kind='bar', stacked=True, color = ['black', 'lightgreen'])
plt.xlabel('Alone?')
plt.ylabel('Survived?')
plt.title('Stacked bar plot of Survival by whether passengers were travelling alone')
plt.show()
```

The above stacked bar plot shows the survival status of passengers on the titanic (percentages) who were either traveling alone or with others. In this case, a surivival status of "0" means that the passenger did not survive, while a status of "1" means that they did survive. The graph demonstrates that approximately 70% of passengers who were travelling alone did not survive, while only 50% of passengers who were travelling with others did not survive.

Marks:

Channels: 
- Position (the bars indicating passengers who surived are situated above the bars indicating passengers who died. Also, the position on the x axis represents whether passengers were alone)
- Size (the height of the bar indicates the proportion (0 to 1) of passengers in each category
- Color (green indicates passengers who survived, and black indicates passengers who did not survive

```python
# remove null values from the age column
titanic_data['age'].unique()
titanic_data['age'] = titanic_data['age'].dropna()
```

```python
sns.kdeplot(x = 'age', hue = 'survived', data=titanic_data, common_norm = False)
plt.xlabel('Age of Passengers')
plt.ylabel('Density')
plt.title('Age by Survival in Titanic Passengers')
plt.show()
```

The above density plot compares the distributions of age among passengers who did and did not survive on the titanic. The blue line represents passengers who did not survive; the mode is approximately 22 years old, and there is a slight positive skew. The orange line represents passengers who did survive; the general shape of the KDE distribution resembles that of the blue line, but the mode age is slightly older at 30 years old.

```python
titanic_data['survived'] = titanic_data['survived'].astype('str')

sns.countplot(x='pclass', hue = 'survived', data=titanic_data, palette={'0': 'black', '1': 'lightgreen'})
plt.xlabel('Passenger Class')
plt.ylabel('Number Survived?')
plt.title('Side-By-Side Bar Plot for Passenger Class and Survival')
plt.show()
```

The above side-by-side bar plot shows the number of passengers who died and survived in each of the three passenger class categories. A surivival status of "0" means that the passenger did not survive, while a status of "1" means that they did survive. The plot shows that significantly more passengers who were in the bottom class 3 died (>350), while under 100 died in classes 1 and 2. The numbers of people who survived in each of the three classes ranged from around 90-140, so there was not as significant a trend among the different classes.
